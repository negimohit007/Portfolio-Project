```python
from tkinter import *
from tkinter import messagebox
from tkinter.ttk import Combobox
from tkinter import Toplevel
```

```python
# Form-1
root = Tk()
root.title("Registration Form")
root.geometry("600x600")
head = Label(root,text="Course Registration Form",font=("Arial",55,"bold"),fg="white",bg="blue")
head.place(x=300, y=60)
# Full Name
sub1 = Label(root,text="Full Name",font=("Calibri",27,"bold"))
sub1.place(x=350,y=180)
entry1 = Text(root,width=40,height=2,font=("Arial"))
entry1.place(x=600,y=185)
# Email
sub2 = Label(root,text="Email",font=("Calibri",27,"bold"))
sub2.place(x=350,y=255)
entry2 = Text(root,width=40,height=2,font=("Arial"))
entry2.place(x=600,y=260)
# Date of Birth
sub3 = Label(root,text="Date of Birth",font=("Calibri",27,"bold"))
sub3.place(x=350,y=330)
entry3 = Text(root,width=40,height=2,font=("Arial"))
entry3.place(x=600,y=335)
# Format line
subhead = Label(root,text="(Format: dd-mm-yyyy)",font=("Calibri",14,"bold"),fg="blue")
subhead.place(x=350,y=370)
# Gender
sub4 = Label(root,text="Gender",font=("Calibri",27,"bold"))
sub4.place(x=350,y=410)
x = IntVar()
r1 = Radiobutton(root,text="Male",variable=x,value=1,font=("Arial"))
r1.place(x=600,y=420)
r2 = Radiobutton(root,text="Female",variable=x,value=2,font=("Arial"))
r2.place(x=700,y=420)
# contact
sub5 = Label(root,text="Contact",font=("Calibri",27,"bold"))
sub5.place(x=350,y=480)
entry3 = Text(root,width=40,height=2,font=("Arial"))
entry3.place(x=600,y=480)
# Select Course
sub6 = Label(root,text="Select Course",font=("Calibri",27,"bold"))
sub6.place(x=350,y=560)
# combobox
cb = Combobox(root,values=['Web Develpment','Data Science','Cloud Computing','Data Analytics',
                          'Digital Marketing','Cyber Security','AWS','Java','Python'],font=('Arial'),
                          width=25,height=2)
cb.set('Select course')
cb.place(x=600,y=570)
# Submit
def test():
    messagebox.showinfo("Form","Successful Submit")
sub = Button(root,text="Submit",font=("Calibri",18,"bold"),fg="white",bg="green",width=15,command=test)
sub.place(x=480,y=640)
# cancel
canc = Button(root,text="Cancel",font=("Calibri",18,"bold"),fg="white",bg="red",command=root.destroy,width=15)
canc.place(x=840,y=640)
root.mainloop()
```

```python
#Form-2
root = Tk()
root.title("Account Login")
root.geometry("600x600")

# separte window for login
def win_login():
    login_window = Toplevel(root)
    login_window.title("Account Login")
    login_window.geometry("400x300")
    
    Label(login_window,text = "Email:",font=("Arial",14)).place(x=50,y=50)
    email = Entry(login_window,font=("Arial",14))
    email.place(x=150,y=50)

    Label(login_window,text="Password:",font=("Arial",14)).place(x=50,y=150)
    password = Entry(login_window,font=("Arial",14),show="*")
    password.place(x=150,y=150)

    def login_action():
        if email.get() and password.get():
            messagebox.showinfo("Login","Login Successful")
            login_window.destroy()
        else:
            messagebox.showerror("Error","Please fill in all fields")
    Button(login_window, text="Submit", font=("Arial", 14), command=login_action).place(x=150, y=200)
    
# fucntion for register   

def window_register():
    register_window = Toplevel(root)
    register_window.title("Register")
    register_window.geometry("600x600")

    Label(register_window,text = "Enter Name:",font=("Arial",14)).place(x=50,y=50)
    name = Entry(register_window,font=("Arial",14))
    name.place(x=210,y=50)

    Label(register_window,text="Enter Email:",font=("Arial",14)).place(x=50,y=100)
    email = Entry(register_window,font=("Arial",14))
    email.place(x=210,y=100)

    Label(register_window,text="Contact:",font=("Arial",14)).place(x=50,y=150)
    contact = Entry(register_window,font=("Arial",14))
    contact.place(x=210,y=150)

    Label(register_window,text="Select Country:",font=("Arial",14)).place(x=50,y=200)
    country = Combobox(register_window,values=["India","USA","Austrila","South Africa","Germany", 
                                "Italy","South Arabia","England","France"],font=("Arial",14))
    country.set("India")
    country.place(x=210,y=200)

    Label(register_window,text="Enter Password:",font=("Arial",14)).place(x=50,y=250)
    password = Entry(register_window,font=("Arial",14))
    password.place(x=210,y=250)

    def register_action():
        if name.get() and email.get() and contact.get() and country.get() and password.get():
            messagebox.showinfo("Register","Successful Register")
            register_window.destroy()
        else:
            messagebox.showerror("Error Show","Please fill all the fields")
            
    Button(register_window,text="Submit",font=("Arial",18),command=register_action).place(x=180,y=350)
            
head = Label(root,text="Select Your Choice",font=("Arial",55,"bold"),fg="white",bg="blue")
head.place(x=300, y=60)
login = Button(root,text = "Login",font=("Arial",45,"bold"),command=win_login)
login.place(x=500,y=200)   
register = Button(root,text = "Register",font=("Arial",45,"bold"),command=window_register)
register.place(x=450,y=380)
root.mainloop()
```
